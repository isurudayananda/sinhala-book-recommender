## Collaborative Filtering Based Recommender System

```python
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Ratings'] > 0
padhe_likhe_users = x[x].index
x
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book_Title').count()['Ratings']>=1
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book_Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book_Title',columns='User-ID',values='Ratings')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book_Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book_Title')['Book_Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book_Title')['Author'].values))
       
        
        data.append(item)
    
    return data
```

```python
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np

# Sample data
data = {
    'age': ['25-34', '35-44', np.nan, '18-24'],
    'gender': ['Male', 'Female', 'Male', np.nan],
    'rating': [5, 4, np.nan, 3]
}

df = pd.DataFrame(data)

# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns

# Impute numeric columns with mean
imputer_numeric = SimpleImputer(strategy='mean')
df[numeric_cols] = imputer_numeric.fit_transform(df[numeric_cols])

# Impute non-numeric columns with most frequent
imputer_non_numeric = SimpleImputer(strategy='most_frequent')
df[non_numeric_cols] = imputer_non_numeric.fit_transform(df[non_numeric_cols])

print(df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Load the data
ratings = pd.read_csv('dataset/ratings.csv')
books = pd.read_csv('dataset/books.csv')
users = pd.read_csv('dataset/users.csv')

# Preprocess the data
ratings = ratings.dropna()
ratings['Ratings'] = ratings['Ratings'].astype(float)

# Aggregate ratings to ensure unique combinations of User-ID and ISBN
ratings = ratings.groupby(['User-ID', 'ISBN']).agg({'Ratings': 'mean'}).reset_index()

# Create the user-item matrix
user_item_matrix = ratings.pivot(index='User-ID', columns='ISBN', values='Ratings').fillna(0)
user_item_matrix_sparse = csr_matrix(user_item_matrix.values)

# Train the model using Nearest Neighbors
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(user_item_matrix_sparse)

# Function to make book recommendations
def recommend_books(user_id, n_recommendations=5):
    user_index = user_item_matrix.index.get_loc(user_id)
    distances, indices = model_knn.kneighbors(user_item_matrix.iloc[user_index, :].values.reshape(1, -1), n_neighbors=n_recommendations+1)
    recommendations = []
    for i in range(1, len(distances.flatten())):
        book_index = user_item_matrix.columns[indices.flatten()[i]]
        book_title = books[books['ISBN'] == book_index]['Book_Title'].values[0]
        recommendations.append(book_title)
    return recommendations

# Example usage
user_id = 1
print(f"Recommendations for user {user_id}: {recommend_books(user_id)}")
```

```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# Load the data
ratings = pd.read_csv('dataset/ratings.csv')

# Prepare the data for Surprise
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings[['User-ID', 'ISBN', 'Ratings']], reader)

# Split the data into training and test sets
trainset, testset = train_test_split(data, test_size=0.25)

# Train the SVD model
model = SVD()
model.fit(trainset)
```

```python
# Evaluate the model
predictions = model.test(testset)
accuracy.rmse(predictions)
```
