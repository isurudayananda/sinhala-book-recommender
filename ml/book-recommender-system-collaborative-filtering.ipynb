## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('ISBN').count()['Ratings'].reset_index()
num_rating_df
```

```python
num_rating_df.rename(columns={'Ratings':'num_ratings'},inplace=True)
num_rating_df
```

```python
print(ratings_with_name['Ratings'].unique())
```

```python
ratings_with_name['Ratings'] = pd.to_numeric(ratings_with_name['Ratings'], errors='coerce')
```

```python
avg_rating_df = ratings_with_name.groupby('ISBN')['Ratings'].mean().reset_index()
```

```python
avg_rating_df.rename(columns={'Ratings': 'avg_rating'}, inplace=True)
```

## Collaborative Filtering Based Recommender System

```python
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Ratings'] > 0
padhe_likhe_users = x[x].index

x
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book_Title').count()['Ratings']>=1
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book_Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book_Title',columns='User-ID',values='Ratings')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book_Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book_Title')['Book_Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book_Title')['Author'].values))
       
        
        data.append(item)
    
    return data
```

```python
recommend('kaluwarayaa')
```
